# Plotting model attention distributions

For test year of 2015.

```python
import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.insert(0,'../code')
from general_utils import standard_model_attention,time_model_attention
```

```python
def plot_attention_mat(att_mat):
    """ Plots attention matrix
    """
    plt.figure(figsize=(8,8))
    plt.imshow(att_mat)
    plt.xlabel("Sites") 
    plt.xticks([i for i in range(0,att_mat.shape[1])],[i for i in range(1,att_mat.shape[1]+1)])
    plt.ylabel("Sites")
    plt.yticks([i for i in range(0,att_mat.shape[0])],[i for i in range(1,att_mat.shape[0]+1)])
    print()
    plt.show()
```

## Standard modeling

### Summer

```python
att_mat,att_mat_p = standard_model_attention("summer","graph_modeling_3_t_2")
```

```python
plot_attention_mat(att_mat)
```

```python
plot_attention_mat(att_mat_p)
```

### Winter

```python
att_mat,att_mat_p = standard_model_attention("winter","graph_modeling_3_t_w_2")
```

```python
plot_attention_mat(att_mat)
```

```python
plot_attention_mat(att_mat_p)
```
